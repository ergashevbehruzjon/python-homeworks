```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 400)
y = x**2 - 4*x + 4

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$f(x) = x^2 - 4x + 4$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x^2 - 4x + 4')
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 400)
sin_y = np.sin(x)
cos_y = np.cos(x)

plt.figure(figsize=(8, 6))
plt.plot(x, sin_y, label='sin(x)', linestyle='-', marker='o', color='b')
plt.plot(x, cos_y, label='cos(x)', linestyle='--', marker='x', color='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of sin(x) and cos(x)')
plt.legend()
plt.show()
```

```python
x = np.linspace(-10, 10, 400)
x_positive = np.linspace(0, 10, 400)

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].plot(x, x**3, color='b')
axs[0, 0].set_title('f(x) = x^3')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('f(x)')

axs[0, 1].plot(x, np.sin(x), color='g')
axs[0, 1].set_title('f(x) = sin(x)')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('f(x)')

axs[1, 0].plot(x, np.exp(x), color='r')
axs[1, 0].set_title('f(x) = e^x')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('f(x)')

axs[1, 1].plot(x_positive, np.log(x_positive + 1), color='m')
axs[1, 1].set_title('f(x) = log(x+1)')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('f(x)')

plt.show()
```

```python
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, c='b', marker='o')
plt.title('Scatter Plot of 100 Random Points')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
data = np.random.normal(0, 1, 1000)

plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, alpha=0.75, color='g')
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.cos(x**2 + y**2)

fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis')
fig.colorbar(surf)
plt.title('3D Surface Plot of f(x, y) = cos(x^2 + y^2)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```

```python
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [200, 150, 250, 175, 225]

plt.figure(figsize=(8, 6))
plt.bar(products, sales, color=['b', 'g', 'r', 'c', 'm'])
plt.title('Sales Data for Five Products')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

```python
categories = ['Category A', 'Category B', 'Category C']
time_periods = ['T1', 'T2', 'T3', 'T4']
data = np.random.randint(1, 10, (3, 4))

bar_width = 0.5
plt.bar(time_periods, data[0], bar_width, label=categories[0])
plt.bar(time_periods, data[1], bar_width, bottom=data[0], label=categories[1])
plt.bar(time_periods, data[2], bar_width, bottom=data[0] + data[1], label=categories[2])

plt.title('Stacked Bar Chart')
plt.xlabel('Time Periods')
plt.ylabel('Values')
plt.legend()

plt.show()
```
